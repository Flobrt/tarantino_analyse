```python
# Je créé une nouvelle colonnes qui me permet de distinguer si un mot est une insulte ou une mort

def detect_words(values, dictionary):
    result = []
    for lib in values:
        operation_type = "x"
        for word, val in dictionary.items():
            if word in lib:
                operation_type = val
        result.append(operation_type)
    return result
data["jurons"] = detect_words(data["mot"], JURONS)
    




# Et supprimes la colonne inutile

for c in data.columns:
    if c not in ['film','mot','minutes_simples','jurons']:
        del data[c]

print(data)
```

```python
# Je regarde si le nombre de grossièrtés dans les films est proportionnelle au nombre de mort 
data_crosstab = pd.crosstab(data.film, data.jurons, margins = False)

data_crosstab['ratio']= data_crosstab['insult']/data_crosstab['mort']

print(data_crosstab)
```

```python
df2 = data[["film", "minutes_simples", "jurons"]]
```

```python
# j'écris une fonction qui me fournit le diagramme pour un film choisit, de nombre de mort et d'injure par minutes
def graphic_movie(film):

  print(film)

  # je recupère juste les données dont j'ai besoin pour mon diagramme et reinitialise l'index de chacunes
  df_film = df2 [df2['film'] == film]
  df_film.reset_index(drop = True, inplace = True)

  # je nomme mes 2 colonnes pour les utilisées plus tard
  insulte = df_film['jurons']
  data2 = df_film['minutes_simples']

  # je créer mon dictionnaire
  res = { key : [insulte[idx]
      for idx in range(len(insulte)) if data2[idx]== key] 
      for key in set(data2)} 

  # je créer 3 dictionnaires vides
  A = []
  B = []
  C = []

  for cle in res.keys():
    A.append(cle)



  for valeur in res.values():
  
    B.append(valeur.count('insult'))
    C.append(valeur.count('mort'))
 

  # je créer mon graphique
  plt.bar(A, B)
  plt.bar(A, C, bottom = np.array(B), color = "red")

  plt.ylim([0, 20])
  
  
```
